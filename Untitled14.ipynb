```python
# Зубчук Владислав,8 група,7 варіант
##Метод Ейлора
import numpy as np 
import matplotlib.pyplot as plt  

def f(x, y): 
    return x + np.cos(y/np.e)  
a = 1.4 # ліва межа відрізку 
b = 2.4 # права межа відрізку 
h = 0.2 # крок 
y0 = 2.5 # початкова умова 
n = int((b - a) / h) # кількість кроків  
x = np.array ([i*h for i in range(0, 6 )]) # задаємо x генератором списків 
y=np.empty(n+1)  
y[0] = y0 
for i in range (n): 
    y[i+1]=y[i]+f(x[i],y[i])*h  
y_rounded=np.round_(y,4)  
print("x=",x, "\ny=",y_rounded)  
plt.plot(x,y, "o",x,y, "red")  
plt.xlabel("x")  
plt.ylabel("y")  
plt.title("Метод Ейлера")  
plt.legend(["точки", "x + np.sin(y/sqrt(0.7))"])  
plt.grid()  
plt.show()
```

```python
#Метод Ейлера-Коші 
import numpy as np 

import matplotlib.pyplot as plt  
def f(x, y): 
    return x + np.sin(y/np.sqrt(7))  
a = 0.5 # ліва межа відрізку 
b = 1.5 # права межа відрізку 
h = 0.2 # крок 
y0 = 0.6 # початкова умова 
n = int((b - a) / h) # кількість кроків 
x = np.array ([i*h for i in range(0, 6 )]) # задаємо x генератором списків 
y=np.empty(n+1)  
y[0] = y0 
for i in range(n):   
     y[i+1]=y[i]+(f(x[i],y[i])+f(x[i+1],y[i])+h*f(x[i],y[i]))*h/2  
y_rounded=np.round_(y,4) 
print("x=",x, "\ny=",y_rounded)  
plt.plot(x,y, "o",x,y, "red")  
plt.xlabel("x")  
plt.ylabel("y")  
plt.title("Метод Ейлера-Коші")  
plt.legend(["точки", "x + np.sin(y/sqrt(11)"])  
plt.grid()  
plt.show()
```

```python
#Розв’язок диф. Рівняння за допомогою scipy.integrate import odeint 

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt  
# function that returns dy/dx 
def model(y,x): 
    x + np.sin(y/np.sqrt(7)) 
    return x + np.sin(y/np.sqrt(11))  
# initial condition 
y0 = 0.6
# x points 
x = np.array ([i*0.2 for i in range(0, 6 )]) 
# solve ODE 
y = odeint(model,y0,x) 
print('x=', x) 
print('y=',y)  
# plot results 
plt.plot(x,y) 
plt.xlabel('x') 
plt.ylabel('y(x)') 
plt.title('Розв’язання диф. рівняння') 
plt.grid()  
plt.show()
```
